# Thompson Sampling

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

## Implementing Thompson Sampling

```
# numbers_of_rewards_1 = [0] * d
# numbers_of_rewards_0 = [0] * d so basically in this we are giving two rewards in the round n for selection 1 & 0

# max_random = in this we will select the add which has the max highest random value ML-A-Z-pg-411/ highest

# step:3
# we have to select the add which has the highest value and that we can get by that i for loop
# thats why we are saying add = i

# n = rows
# ad = after computation what we will get the value for the add column
# in step 1 our problem is stating that in a particular row n and  add from our dataset values if the value is 1
# then reward == 1  then change or increment the value for numbers_of_rewards_1
#  or
#  from our dataset values if the value is 0 then reward == 0 then change or increment the value for numbers_of_rewards_0
#  or
```

```
import random
N = 10000
d = 10
ads_selected = []
# Ni1
numbers_of_rewards_1 = [0] * d
# Ni0
numbers_of_rewards_0 = [0] * d
total_reward = 0
for n in range(0, N):
  ad = 0
  max_random = 0
  for i in range(0, d):
    random_beta = random.betavariate(numbers_of_rewards_1[i] + 1, numbers_of_rewards_0[i] + 1)
    if (random_beta > max_random):
      max_random = random_beta
      ad = i
  ads_selected.append(ad)
  reward = dataset.values[n, ad]
  if reward == 1:
    numbers_of_rewards_1[ad] = numbers_of_rewards_1[ad] + 1
  else:
    numbers_of_rewards_0[ad] = numbers_of_rewards_0[ad] + 1
  total_reward = total_reward + reward
```

## Visualising the results - Histogram

```
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
